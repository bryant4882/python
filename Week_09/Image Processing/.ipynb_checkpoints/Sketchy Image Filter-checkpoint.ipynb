### There are four steps to creating a sketchy image filter:

1. Convert the image to grayscale
2. Invert it
3. Blur the inverted image
4. Dodge blend the grayscale image and the blurred image

```python
#importing required modules
import cv2
import common
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
import pylab
pylab.rcParams["figure.figsize"] = (10.0,8.0)
```

```python
#base image
img = cv2.imread("Images/2.jpg")
```

```python
#step 1: converting to grayscale image

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap="gray")
```

```python
#Step 2: Inverting it

img_invert = 255 - img_gray

#or use opencv function called bitwise_not()

inverted_img = cv2.bitwise_not(img_gray)
plt.imshow(inverted_img)
```

```python
#step 3: Blurring the image using gaussian blur

img_blur = cv2.GaussianBlur(inverted_img,(15,15),10)
plt.imshow(img_blur)
```

```python
#step 4: Dodging and blending the two images
#we use the arithmetic operation of dividing two images

def dodge(image,mask):
    return cv2.divide(image,255-mask,scale=256)
```

```python
sketchy_image = dodge(img_gray,img_blur)
plt.imshow(sketchy_image, cmap="gray")
```
