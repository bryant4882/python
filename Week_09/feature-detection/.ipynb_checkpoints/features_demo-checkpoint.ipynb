### Feature Detection #1
__Canny Edge Detector__

[Canny](https://docs.opencv.org/2.4/modules/imgproc/doc/feature_detection.html?Canny) finds edges in an image.

In Python, the standard call for running Canny looks like this:

`cv2.Canny(image, threshold1, threshold2[, edges[, apertureSize[, L2gradient]]])`

We are going to focus on manipulating the first three parameters for this function.

The first parameter `image` takes in the image we want to manipulate.

The second and third parameters, `threshold1` and `threshold2`, refer to the thresholder used in the Canny operator called hysteresis. In image processing, hysteresis compares two images to build an intermediate image (you can do more research on this if this is something interesting). We can manipulate these values to see what gives us the best result.

```python
# Detecting edges using Canny edge Detection

#color correct the original image
input_image_1 = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
```

```python
#calling in Canny
edges = cv2.Canny(input_image_1, 100, 200)

#plotting the original image and output image side by side
#original image
plt.subplot(121),plt.imshow(input_image_1, cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
#Canny edge image
plt.subplot(122),plt.imshow(edges, cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
#plot both subplots
plt.show()
```

### Feature Detection #2
__Harris Corner Detector__

[Harris Corner Detector](https://docs.opencv.org/2.4/modules/imgproc/doc/feature_detection.html?highlight=cornerharris) is often used in CV to extract corners to infer features of an image.

<img src="images/harris.png" width="400"/>

In Python, the standard call for running a Harris Corner Detector looks like this:

`cv2.cornerHarris(src, blockSize, ksize, k[, dst[, borderType]])`

We are going to focus on manipulating the four different parameters for this function.

The first parameter `src` is the input image.

The second parameter `blockSize` is the Neighborhood size. For every pixel, the function will consider `blockSize * blockSize` neighborhood and will calculate the covariation matrix of derivatives over that neighborhood.

The third parameter, `ksize` is the Sobel operator aperature size. The Sobel operator, is used in edge detection algorithms to creates an image emphasizing edges.

The fourth parameter is the Harris Threshold. The detector will find points where the value of the calculated corner response function is higher than the set threshold.

The returned result is usually stored in the variable `dst`. `dst` refers to the image that stores the Harris detector responses.

```python
#using the harris corner detection function to find the corners in our image

#making a copy of our original image and storing it in our variable
harris_test=input_image.copy()

#grayscaling the image (very common in image processing)
#this will return a numpy ndarray with the grayscale values
gray = cv2.cvtColor(harris_test,cv2.COLOR_BGR2GRAY)

#you can also change the values into floating points
#sometimes this is required
gray = np.float32(gray)

#show output
gray
```

```python
#we are going to define the blockSize, ksize, and threshold
#the sobel kernel works best when it's fairly small
#you can play around with the values to see which combination detects the corners the best
blocksize=4
kernel_size=3 
threshold = 0.05

#calling the detector & storing our detection result in dst
dst = cv2.cornerHarris(gray, blocksize, kernel_size, threshold) 
#result is dilated for marking the corners, this is visualisation related and just makes them bigger
dst = cv2.dilate(dst, None)
#we then plot these on the input image for visualisation purposes
harris_test[dst > 0.01 * dst.max()] = [100,100,255]
#plotting our image using matplotlib & imshow
plt.imshow(cv2.cvtColor(harris_test, cv2.COLOR_BGR2RGB))
```

### Feature Detection #3
__ORB (oriented BRIEF)__

[ORB](https://docs.opencv.org/3.4/db/d95/classcv_1_1ORB.html#details) keypoint detector and descriptor extractor helps us detect stable keypoints in our image. ORB is built on other feature detection algorithms (FAST) and was mainly developed because the other two popular keypoint detection algorithms, SIFT and SURF, are patented.

In Python, the standard call for running ORB looks like this:

`cv2.ORB_create([, nfeatures[, scaleFactor[, nlevels[, edgeThreshold[, firstLevel[, WTA_K[, scoreType[, patchSize[, fastThreshold]]]]]]]]])`

The parameters here are optional. For this notebook, we won't be manipulating any of the parameters.

You can read more about ORB [here](https://medium.com/data-breach/introduction-to-orb-oriented-fast-and-rotated-brief-4220e8ec40cf) (both images are from this article).

```python
# finding keypoints/points of interest using ORB

orbimg = input_image.copy()

orb = cv2.ORB_create()
#find the keypoints with ORB
kp = orb.detect(orbimg,None)
#compute the descriptors with ORB
kp, des = orb.compute(orbimg, kp)
#draw keypoints
cv2.drawKeypoints(orbimg,kp,orbimg)
#plotting our image using matplotlib & imshow
plt.imshow(cv2.cvtColor(orbimg, cv2.COLOR_BGR2RGB))
```
